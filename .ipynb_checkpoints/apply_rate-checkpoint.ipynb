```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from scipy import stats
from sklearn.cluster import KMeans
from datetime import datetime
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
```

# NEW data

```python
data = pd.read_csv('Apply_Rate_2019.csv')
```

```python
data.head()
```

```python
data.search_date_pacific=pd.to_datetime(data.search_date_pacific, format='%Y-%m-%d')
data.search_date_pacific = data.search_date_pacific.map(lambda x: x.date())
```

```python
data.info()
```

```python
data.describe()
```

```python
data.describe(include=object)
```

```python
data.title_proximity_tfidf.fillna(data.title_proximity_tfidf.median(), inplace=True)
data.description_proximity_tfidf.fillna(data.description_proximity_tfidf.median(), inplace=True)
data.city_match.fillna(data.city_match.max(), inplace=True)
```

```python
plt.bar(['0','1'], data['apply'].value_counts())
```

```python
c = data.corr()
```

```python
sns.heatmap(c)
```

```python
clmns = ['title_proximity_tfidf', 'description_proximity_tfidf',
       'main_query_tfidf', 'query_jl_score', 'query_title_score', 'city_match',
       'job_age_days', 'apply']
fig = plt.figure(figsize=(200,200))
for i,c in enumerate(clmns):
    ax = fig.add_subplot(3,3,i+1,)
    ax.set_title(c, fontsize=120)
    data[c].hist(bins=20)
```

```python
X= data[['title_proximity_tfidf','description_proximity_tfidf','main_query_tfidf','query_jl_score','query_title_score','city_match','job_age_days']]
Y = data[['apply']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3)
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(X_train, y_train)
```

```python
mnb.score(X_test, y_test)
```

```python
dt = DecisionTreeClassifier(max_depth=4).fit(X_train, y_train)
```

```python
dt.score(X_test, y_test)
```

```python
rf = RandomForestClassifier(n_estimators=20, max_depth=4,min_samples_split=4).fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

```python
bc = BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth=4)).fit(X_train,y_train)
```

```python
bc.score(X_test, y_test)
```

```python
adaboost= AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=6),algorithm='SAMME').fit(X_train, y_train)
```

```python
adaboost.score(X_test, y_test)
```

```python
svm = SVC().fit(X_train, y_train)
```
